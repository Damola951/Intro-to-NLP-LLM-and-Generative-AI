# Mini Text-Based Attention Visualizer
This notebook includes a simple attention visualizer that shows how much attention weight each word receives in a sentence. The design uses plain text bars to illustrate attention strength without any graphics.


## Design Explanation
- The visualizer uses simple text and bars so it works in any notebook environment without graphics.
- Tokens are cleaned and lowercased to match the notebook's tokenization concept.
- We normalize by the maximum weight so the strongest word always fills the bar, making relative attention easier to compare.
- The function supports both user-provided weights and a default illustrative pattern.


```python
def visualize_attention(sentence, attention_weights=None, width=40):
    """Visualize attention weights for each token in a sentence.

    Design choices:
    - Simple whitespace tokenization and punctuation stripping keeps the visualizer easy to understand.
    - Optional attention weights let the function support both real model output and illustrative examples.
    - Fixed-width bars show relative strength clearly without graphics.
    - Normalizing by the maximum weight makes the strongest word fill the bar and keeps comparisons consistent.
    """

    # Tokenize the sentence by splitting on whitespace and strip common punctuation.
    tokens = [token.strip('.,!?;:\"\'').lower() for token in sentence.split()]
    tokens = [token for token in tokens if token]

    # If no explicit attention weights are provided, build a simple illustrative pattern.
    if attention_weights is None:
        attention_weights = [1.0 / len(tokens)] * len(tokens)
        if tokens:
            attention_weights[-1] = 0.5
            if len(tokens) > 1:
                attention_weights[0] = 0.3
            total = sum(attention_weights)
            attention_weights = [w / total for w in attention_weights]

    # Ensure the weights list matches the number of tokens.
    if len(attention_weights) != len(tokens):
        raise ValueError('Number of attention weights must match the number of tokens.')

    # Normalize by the maximum weight for display scaling.
    max_weight = max(attention_weights) if attention_weights else 1.0
    normalized = [w / max_weight for w in attention_weights]

    print('Attention Visualizer (text-based)')
    print('-' * (width + 30))
    for token, weight, norm in zip(tokens, attention_weights, normalized):
        bar_length = int(norm * width)
        bar = '█' * bar_length + ' ' * (width - bar_length)
        print(f'{token:15s} | {weight:0.3f} | {bar}')
    print('-' * (width + 30))


# Example usage with the notebook attention example.
sentence = 'The bank is by the river'
weights = [0.05, 0.20, 0.10, 0.10, 0.10, 0.45]
visualize_attention(sentence, weights)
```

```python
# Simple rule-based chatbot with keyword matching.

faq_rules = {
    'hello': 'Hello! I am a simple chatbot. How can I help you today?',
    'hi': 'Hi there! Ask me about NLP, transformers, or how attention works.',
    'help': 'You can ask about the course, NLP concepts, or say "example" for a sample query.',
    'attention': 'Attention helps models focus on relevant words. It assigns weights to each input token.',
    'transformer': 'Transformers use self-attention to process all tokens in parallel and understand context.',
    'nlp': 'NLP stands for Natural Language Processing, which lets computers understand text and speech.',
    'bye': 'Goodbye! Feel free to ask more questions anytime. And don\'t forget to submit me on Jun 15th.',
    'thanks': 'You are welcome! Glad I could help.',
    'thank': 'You are welcome! Glad I could help.',
    'okay': 'Okay! I will remember that and I am here if you need anything else.'
}

# The chatbot function is designed to match keywords rather than full sentences.
# This is a rule-based design because it does not learn from data; it only follows explicit instructions.
# It prioritizes exact keyword matches and falls back to a default message when no keyword matches.

def rule_based_chatbot(user_input):
    """Return a response based on keyword matching in the user's input."""
    cleaned = user_input.lower()
    # Check each keyword in the user input and return the first matching response.
    for keyword, response in faq_rules.items():
        if keyword in cleaned:
            return response
    return "Sorry, I don't know the answer to that yet. Try asking about attention, transformers, or NLP."

# Example interaction with the chatbot.
user_examples = [
    'Hello',
    'Can you explain attention?',
    'What is a transformer?',
    'Tell me about NLP',
    'Thanks for your help',
    'Goodbye',
    'Okay'
]

for text in user_examples:
    print('User:', text)
    print('Bot: ', rule_based_chatbot(text))
    print()
```

```python
# Interactive chatbot helper for manual input.

def chat_with_bot():
    """Run a simple input loop that lets the user talk to the rule-based chatbot."""
    print('Type a message to the chatbot, or type "exit" to quit.')
    while True:
        user_input = input('You: ').strip()
        if not user_input:
            print('Please type something or enter "exit" to quit.')
            continue
        if user_input.lower() in {'exit', 'quit', 'bye'}:
            print('Bot: Goodbye!')
            break
        response = rule_based_chatbot(user_input)
        print('Bot:', response)

# Auto-run demonstration of the chatbot conversation.
demo_inputs = [
    'Hello',
    'Can you explain attention?',
    'What is a transformer?',
    'Tell me about NLP',
    'Thanks for your help',
    'Goodbye'
]

print('Auto-run chatbot demo:')
for text in demo_inputs:
    print('User:', text)
    print('Bot: ', rule_based_chatbot(text))
    print()

# If you want interactive mode, uncomment the line below and run this cell again.
# chat_with_bot()
```
